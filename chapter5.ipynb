# 5장 유사성과 모호성

## 5.1 단어의 의미(word sense)

### 5.1.1 단어의 의미의 관계
* 단어는 표제어(lemma)(겉으로 보이는 형태)안에 여러 의미를 담고 있다.
* 대부분의 사람은 주변 정보(context)에 따라서 숨겨진 의미를 파악하고 이해한다.
* 주변 정보가 부적하여 모호성이 증가한 경우, 사람이라 해도 본래의 의미를 잘못 이해하기도 한다.

<그림>

* 단어는 그 형태는 공유하더라도 서로 다른 뜻을 가진 의미로 구성될 수 있다.
* 한 가진 형태의 단어에 여러 의미가 포함되어 생기는 '중의성' 문제는 자연어 처리에 배우 중요한 위치를 차지한다. 
* 결국 인공지능은, 바깥의 '단어'라는 형태의 매개체를 이해하고 안쪽의 '의미'로 변환하여 사용해야 한다.

### 5.1.2 동형어와 다의어
* 동형어(homonym) : 형태는 같으나 뜻이 서로 다른 단어
* 다의어(polysemy)
* 동형어와 다의어의 차이 : 다의어의 경우 한 형태의 단어가 여러 의미를 지니면서도 그 의미들이 서로 관련이 있는 뜻을 가진다. 동형어는 아예 어원이 다른 의미들이 같은 형태를 띠는 단어이다.

<그림>

* 한 형태 내에서 여러 의미를 지니는 동형어 또는 다의어의 경우 => 단어 중의성 해소(word-sense disambiguation)(WSD) 방법을 통해 단어의 의미를 더 명확히 하는 과정이 필요하다.

### 5.1.3
* 다른 형태의 단어와 의미가 서로 겹치는 경우가 발생할 수 있다.
* 동의어(synonym) : 다른 형태이지만 의미가 같은 단어
* 의미라는 것이 완전히 동일하지 않을 수 있지만 어느 정도 비슷할 것이라는 일종의 동의(consensus)가 있을 것이다.
* 동의어가 여러 개 존재할 때 이들의 집합을 동의어 집합(synset)이라고 부른다.

<그림>

### 5.1.4 상위어와 하위어
* 사람이 사용하는 단어는 하나의 추상적 개념을 나타내는데, 그 개념들을 포함하는 상위 개념 또는 하위 개념이 존재할수 있다.
* 상위 개념을 가리키는 단어를 상위어(hypernym)라고 하고, 하위 개념을 표현하는 단어를 하위어(hyponym)이라고 한다.

<그림>

* 이러한 단어들의 어휘 분류(taxonomy)에 따라 단어 간 관계 구조를 계층화할 수 있다.

### 5.1.5 모호성 해소
* 컴퓨터가 가진 것은 텍스트뿐이므로, 그 텍스트가 내포한 진짜 의미를 파악하는 과정이 필요하다. 단어의 겉 형태인 텍스트만으로는 모호성이 매우 높다. 단어가 가지는 모호성을 제거하는 과정을 단어 중의성 해소(WSD)라고 한다.


## 5.2 원핫 인코딩

* 단어를 컴퓨터가 인지할 수 있는 수치로 바꾸는 가장 간단한 방법은 벡터로 표현하는 것이다.
* 그 중에서 가장 기본적인 방법은 원핫 인코딩(one-hot encoding) 방식이다
* 단 하나의 1과 나머지 수많음 0들로 표현된 인코딩 방식을 의미한다.
* 원핫 인코딩 벡터의 차원은 보통 전체 어휘(vocabulary)의 개수가 되며, 당연히 그 숫자는 매우 크다.

* 단어는 불연속적인 심볼로써 이산 확률 변수로 나타낸다. 이산 확률 변수는 그 값을 불연속으로 가질 수밖에 없다. -> 원핫 벡터는 이산 확률 분포로부터 뽑아낸 샘플이라고 할 수 있다.

<그림>

* 사전(dictionary)내의 각 단어를 원핫 인코딩 방식을 통해 벡터로 나타낼 수 있다. but 문제점이 있다! => 벡터의 차원이 너무 커졌다는 것이다.
* 벡터의 많은 부분이 0으로 채워진 벡터를 희소 벡터(sparse vector)라고 한다. 회소 벡터의 가장 큰 문제점은 벡터 간 연산을 할 때 결과값이 0이 된다는 것이다 => 즉, 서로 직교(orthogonal)하는 경우가 많아진다.
(추가설명) 원 핫 인코딩을 취하면 사전의 단어 개수의 차원을 가지는 벡터가 생성된다. 각각의 축에서 1의 값을 가지고 있는 것이기 때문에 결국 어떠한 원 핫 벡터를 곱하든 0이된다. (직교하기 때문에)

### 5.2.1 차원의 저주
* 회소 벡터는 굉장히 큰 차원에 적은 정보만 가지고 있다. 이는 머신러닝에 큰 장벽이 된다. 큰 차원으로 정보를 표현한 것과 작은 차원으로 정보를 표현한 것을 비교하면 같은 크기의 공간에 표현되는 정보가 다를 것이다. 큰 차원으로 정보를 표현했다면 같은 크기로 공간을 나눴을 때 공간이 비어있는 경우가 많을 것이다.

<그림>

* 차원이 늘어날수록 이와 같은 문제가 지수적(exponential) 늘어난다. 이러한 문제를 차원의 저주(curse of dimensionality)라고 부른다. 따라서 차원의 저주를 피하고자 차원을 축소하여 단어를 표현해야 한다.
